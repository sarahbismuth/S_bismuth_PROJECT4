<h5>Feature Importance</h5>

```python
# Feature Importance
import scipy
alpha = 0.05
X = df.drop(['class'], axis=1)
values = {}
for col in X.columns.values:
        A, B = df[col], df["class"]
        dfObserved = pd.crosstab(A, B) 
        chi2, p, dof, expected = scipy.stats.chi2_contingency(dfObserved.values)
        values[col] = p
        if p < alpha:
            # Reject null hypothesis
            print("{} is important. (p = {})".format(col, p))
        else:
            # Accept null hypothesis
            print("{} is NOT important. (p = {})".format(col, p))
```

<h4>Encoding Data</h4>

```python
#as this is textual/character data to train a ML algo we have to encode it.
from sklearn.preprocessing import LabelEncoder #import to encode data

#initalizie the encode object
Enc = LabelEncoder()

#make copy of our data
df_copy = df.copy()

#loop through all columns and transform them
for col in df_copy.columns:
    df_copy[col]=Enc.fit_transform(df[col])
```

```python
#our data has been encoded 
#just for checking we will print our encoded data again.
for col in df_copy.columns:
    print('Column : ',  col)
    print(df_copy[col].unique())
    
#our class column is also encoded p = 1, e = 0 (as per requirements)
```

<h4>Train Test Split</h4>

```python
#preparing our data set for training we will use y for class column and x for all others
# make sure use our encoded data frame
X = df_copy.drop(['class'], axis=1)
Y = df_copy['class']
```

```python
#as per ML algos , we have to split our data in train test split.
# good practice is to split data in 70/30 for train test respectively
from sklearn.model_selection import train_test_split #import train test split for scikit-learn

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state = 42)
```

<h4>Logistic Regression</h4>

```python
#here comes the training of different classifers, as our class lables has only two values
#so its binary classificaiton problem
from sklearn.linear_model import LogisticRegression #import logistic regression model
from sklearn.metrics import accuracy_score #import to check our model score

#make object of our logistic regression
clf = LogisticRegression()
#training
clf.fit(X_train, Y_train)

#making prediction on test data
pred=clf.predict(X_test)

#measure accuracy of test data
lr_acc = round(accuracy_score(Y_test,pred)*100,2)
print('Logistic Regression Accuracy Score is {}%'.format(lr_acc))
```

<h4>KNN</h4>

```python
# Train KNN for our data 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,Y_train)
knn_acc = round(knn.score(X_test,Y_test)*100,2)
print('K-NN accuracy {}%'.format(knn_acc))
```

<h4>SVM</h4>

```python
from sklearn.svm import SVC
svm = SVC(random_state=30, gamma="auto")
svm.fit(X_train,Y_train)
svm_acc = round(svm.score(X_test,Y_test)*100,2)
print("SVM Accracy {}%".format(svm_acc))
```

<h4>NaiveBayes</h4>

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,Y_train)
nb_acc = round(nb.score(X_test,Y_test)*100,2)
print("Gaussian NB accuracy {}%".format(nb_acc))
```
